```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# /home/ 부터 시작
rdd1 = spark.sparkContext.textFile("file:///home/hadoop/jupyter/dataengineering/test.txt")
```

```python
rdd1_2 = rdd1.flatMap(lambda x: x.split(" "))
rdd1_2.collect()
```

```python
rdd1_3 = rdd1_2.map(lambda x : (x, 1))
rdd1_3.collect()
```

```python
rdd1_4 = rdd1_3.reduceByKey(lambda a, b: a+b)
rdd1_4.collect()
```

```python
rdd1_5 = rdd1_4.map(lambda x: (x[1],x[0])).sortByKey()
```

```python
rdd1_5.collect()
```

## RDD 방식은 추상적

## * reduceByKey 동작원리
### 단계 1: 과자 모으기
각 친구가 가져온 과자를 적어본다고 해볼게요:

철수: 초코칩 2개, 사탕 1개
영희: 초코칩 3개, 사탕 2개
민수: 초코칩 1개, 사탕 2개
이렇게 모은 과자 목록을 하나로 합치면:

초코칩 2개 (철수)
사탕 1개 (철수)
초코칩 3개 (영희)
사탕 2개 (영희)
초코칩 1개 (민수)
사탕 2개 (민수)

### 단계 2: 과자 그룹화
이제 같은 종류의 과자를 그룹으로 모아볼게요:

초코칩: 2개, 3개, 1개
사탕: 1개, 2개, 2개
단계 3: 과자 개수 합치기
같은 종류의 과자 개수를 합쳐보아요:

초코칩: 2 + 3 + 1 = 6개
사탕: 1 + 2 + 2 = 5개
결과는:

초코칩: 6개
사탕: 5개
## 이게 바로 reduceByKey가 하는 일이에요! reduceByKey는 같은 키(종류)의 값(과자 개수)을 합쳐주는 역할을 합니다.

```python
data = [
        (("James",None,"Smith"),"OH","M"),
        (("Anna","Rose",""),"NY","F"),
        (("Julia","","Williams"),"OH","F"),
        (("Maria","Anne","Jones"),"NY","M"),
        (("Jen","Mary","Brown"),"NY","M"),
        (("Mike","Mary","Williams"),"OH","M")
        ]
# StructField안에 StructField
from pyspark.sql.types import StructType,StructField, StringType        
schema = StructType([
    StructField('name', StructType([
         StructField('firstname', StringType(), True),
         StructField('middlename', StringType(), True),
         StructField('lastname', StringType(), True)
         ])),
     StructField('state', StringType(), True),
     StructField('gender', StringType(), True)
     ])
df2 = spark.createDataFrame(data = data, schema = schema)
df2.show()
```

```python
simpleData = [("James","Sales","NY",90000,34,10000),
    ("Michael","Sales","NY",86000,56,20000),
    ("Robert","Sales","CA",81000,30,23000),
    ("Maria","Finance","CA",90000,24,23000),
    ("Raman","Finance","CA",99000,40,24000),
    ("Scott","Finance","NY",83000,36,19000),
    ("Jen","Finance","NY",79000,53,15000),
    ("Jeff","Marketing","CA",80000,25,18000),
    ("Kumar","Marketing","NY",91000,50,21000)
  ]

# Create DataFrame
schema = ["employee_name","department","state","salary","age","bonus"]
df = spark.createDataFrame(data=simpleData, schema = schema)
df.printSchema()
df.show(truncate=False)
```

```python
df.withColumn("total_compensation", col("salary") + col("bonus")).show()
```
